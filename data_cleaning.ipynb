```python
import numpy as np
import pandas as pd

df = pd.read_csv("india_youtube_trending _data.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df["publishedAt"] = pd.to_datetime(df["publishedAt"], utc = True)
```

```python
df["trending_date"] = pd.to_datetime(df["trending_date"], utc = True)
```

```python
df["description"] = df["description"].fillna("No Description")
```

```python
df.info()
```

```python
import json

with open("IN_category_id.json", "r") as f:
    data = json.load(f)

cat_dict = {}

for i in data["items"]:
    cat_id = int(i["id"])

    cat_title = i["snippet"]["title"]

    cat_dict[cat_id] = cat_title

print(cat_dict)
```

```python
df["category_name"] = df["categoryId"].map(cat_dict)
```

```python
df.drop(columns=["dislikes"], inplace= True)

print(df.columns)
```

```python
df["title_length"] = df["title"].str.len()

df["engagement_rate"] = (df["likes"] + df["comment_count"] / df["view_count"])
```
